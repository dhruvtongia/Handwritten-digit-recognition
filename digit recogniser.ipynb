```python
import tensorflow as tf
import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Dropout, Activation,Conv2D,MaxPooling2D , Flatten,InputLayer
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("C:\\Users\\Dhruv Tongia\\Downloads\\digit-recognizer\\train.csv",dtype=np.float32)
test_df=pd.read_csv("C:\\Users\\Dhruv Tongia\\Downloads\\digit-recognizer\\test.csv",dtype=np.float32)
```

```python
df
```

```python
df
train=np.array(df.iloc[:,1:]/255)
train=train.reshape(train.shape[0],28,28,1)
label=np.array(df.iloc[:,0])
test=np.array(test_df.iloc[:,0:]/255)
test.shape
test=test.reshape(test.shape[0],28,28,1)
#train_x,test_x,train_y,test_y=train_test_split(train,label,test_size=0.2,random_state=18)
```

```python
label=label.reshape(-1,1)
train.shape
```

```python
model=Sequential()
model.add(InputLayer((28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),activation='relu',padding='valid'))
model.add(MaxPooling2D(strides=(2,2)))
model.add(Conv2D(64,kernel_size=(3,3),activation='relu',padding='valid'))
model.add(MaxPooling2D(strides=(2,2)))
model.add(Conv2D(128,kernel_size=(3,3),activation='relu',padding='valid'))
model.add(MaxPooling2D(strides=(2,2)))
model.add(Dropout(rate=0.5))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(10,activation='softmax'))
model.compile(optmizer='adam',loss='sparse_categorical_crossentropy',metrics=['acc'])

          
```

```python
model.fit(train,label,batch_size=256,epochs=15,validation_split=0.2)
```

```python
ypred=model.predict(test)
ypred=np.argmax(ypred,axis=1)
```

```python
ypred
```

```python
submissions=pd.DataFrame({"ImageId": list(range(1,len(ypred)+1)),
                         "Label": ypred})
submissions.to_csv("DR1.csv", index=False, header=True)
```
